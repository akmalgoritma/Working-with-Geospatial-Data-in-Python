# Coordinate References Systems

Sistem Referensi Koordinat (Coordinate Reference System atau CRS) adalah kerangka kerja yang digunakan dalam ilmu geografi, pemetaan, geospasial, dan sistem informasi geografis (GIS) untuk menentukan dan merepresentasikan posisi atau lokasi objek di permukaan Bumi. Sistem referensi koordinat adalah konsep penting karena membantu kita memetakan, mengukur, dan bekerja dengan data geografis secara akurat dan konsisten. Berikut adalah penjelasan yang lebih komprehensif tentang sistem referensi koordinat:

**1. Tujuan Utama Sistem Referensi Koordinat:** 

Tujuan utama dari sistem referensi koordinat adalah untuk memberikan cara standar dan konsisten untuk menentukan dan merepresentasikan lokasi geografis pada permukaan Bumi. Ini diperlukan untuk berbagai aplikasi, seperti pemetaan, navigasi, analisis geografis, pemodelan iklim, pemantauan lingkungan, manajemen sumber daya alam, dan banyak lagi.

**2. Elemen Utama dalam Sistem Referensi Koordinat:**

Sistem referensi koordinat terdiri dari dua elemen utama: garis lintang (latitude) dan garis bujur (longitude).
- Garis lintang mengukur jarak relatif dari titik ke garis khatulistiwa di utara atau selatan, dengan garis khatulistiwa dianggap sebagai nol derajat lintang.
- Garis bujur mengukur jarak relatif dari titik ke meridian primer (misalnya, Meridian Greenwich) di timur atau barat, dengan meridian utama ini dianggap sebagai nol derajat bujur.


**3. Sistem Koordinat Geografis (Geographic Coordinate System):**

Ini adalah sistem dasar yang menggunakan garis lintang dan garis bujur untuk menentukan lokasi geografis di permukaan Bumi.
Koordinatnya dapat dinyatakan dalam derajat, menit, dan detik (DMS) atau dalam bentuk desimal derajat.



**4. Sistem Koordinat Kartografi (Projected Coordinate System):**

Dalam pemetaan, proyeksi digunakan untuk mengubah representasi tiga dimensi Bumi menjadi representasi dua dimensi pada peta.
Sistem proyeksi ini menggunakan unit seperti meter atau kaki untuk mengukur jarak dan area.





**5. Datum Geodetik:**

Datum geodetik adalah kerangka referensi yang mendefinisikan bentuk dan ukuran Bumi.
Sebagai contoh, beberapa datum yang umum digunakan termasuk WGS 84 (World Geodetic System 1984) dan NAD 83 (North American Datum 1983).

**6. Sistem Lokal (Local Coordinate System):**

Sistem referensi koordinat lokal dapat dibuat untuk proyek atau wilayah tertentu yang tidak tergantung pada garis lintang dan bujur global.
Sistem lokal ini sesuai dengan proyek atau aplikasi tertentu dan mungkin menggunakan koordinat berbasis satuan khusus.

**7. Aplikasi Sistem Referensi Koordinat:**

Sistem referensi koordinat digunakan dalam berbagai aplikasi, termasuk pemetaan, navigasi, analisis geografis, pemodelan iklim, pemantauan lingkungan, perencanaan kota, pemantauan satelit, dan banyak lagi.
Memahami sistem referensi koordinat adalah kunci untuk pemahaman lokasi dan analisis geografis yang akurat. Sistem ini memungkinkan kita untuk bekerja dengan data geografis dalam konteks global atau lokal, serta menghasilkan peta dan analisis yang bermakna.

> - CRS menghubungkan koordinat ke tempat sebenarnya di permukaan bumi
> - Jenis koordinat yang paling dikenal dalam menentukan posisi di bumi adalah koordinat geografis. koordinat ini dinyatakan dengan lintang dan bujur.
> - kemudian bumi yang dalam kenyataannya berbentuk "bola" dapat diubah ke bidang datar 2D melalui proyeksi
> - Dalam setiap proyeksi 3D to 2D akan ada selalu konsekuensi, yaitu berupa distorsi area, sudut atau jarak. Untuk meminimalkan distorsi tersebut berbagai pendekatan digunakan sehingga diperoleh hasil yang optimal pada wilayah itu.
    - proyeksi yang mempertahankan ukuran area geometri > proyeksi albers
    - proyeksi yang mempertahankan sudut > proyeksi mercator

## Specifying a CRS

### Proj4

Proj4 String adalah cara standar untuk mendefinisikan sistem referensi koordinat (Coordinate Reference System atau CRS) dalam bentuk teks. Ini digunakan dalam berbagai perangkat lunak geospasial, termasuk pyproj di Python, untuk mendefinisikan proyeksi peta dan transformasi koordinat. Proj4 String berisi informasi penting tentang proyeksi, datum, unit, dan parameter lain yang diperlukan untuk mengonversi koordinat dari satu sistem referensi ke sistem referensi lainnya.

Sebuah Proj4 String biasanya terdiri dari komponen-komponen berikut:

- +proj: Ini adalah parameter yang menentukan jenis proyeksi peta. Contohnya termasuk merc untuk proyeksi Mercator, utm untuk Universal Transverse Mercator, atau longlat untuk sistem koordinat geografis (latitude/longitude).

- +datum: Ini adalah parameter yang mengidentifikasi datum atau kerangka referensi geodetik yang digunakan. Misalnya, WGS84 untuk World Geodetic System 1984.

- +units: Ini adalah parameter yang mengidentifikasi unit pengukuran. Ini dapat berupa m untuk meter, ft untuk kaki, atau unit lainnya.

- +zone: Parameter ini umumnya digunakan dalam proyeksi Universal Transverse Mercator (UTM) untuk menentukan zona UTM yang digunakan. Misalnya, +zone=33 untuk zona UTM 33.

- +lon_0, +lat_0: Ini adalah parameter yang menentukan titik pusat atau titik asal proyeksi peta. Koordinat longitude dan latitude digunakan untuk menentukan pusat proyeksi.

- +k: Parameter ini mengontrol faktor skala dalam proyeksi. Ini memungkinkan penyesuaian skala yang dapat diterapkan pada proyeksi.

- +x_0, +y_0: Ini adalah parameter yang menentukan offset (pergeseran) dalam koordinat X dan Y. Ini sering digunakan dalam proyeksi yang tidak dimulai di titik nol (0,0).

Contoh Proj4 String untuk proyeksi Mercator adalah:

**+proj=merc +datum=WGS84 +units=m +lon_0=0 +lat_0=0**

Dalam Python, Anda dapat menggunakan pyproj untuk memanfaatkan Proj4 String. Ini memungkinkan Anda untuk mengonversi koordinat antara berbagai sistem referensi dengan mudah. Proj4 String adalah cara yang kuat dan fleksibel untuk mendefinisikan sistem referensi koordinat, dan memahami strukturnya akan membantu Anda berinteraksi dengan berbagai data geografis dalam pemrograman geospasial.

### EPSG Code

EPSG (European Petroleum Survey Group) code adalah sistem identifikasi yang digunakan untuk merujuk dan mengidentifikasi sistem referensi koordinat (Coordinate Reference System atau CRS). Sistem ini dikembangkan oleh European Petroleum Survey Group, yang kemudian diintegrasikan ke dalam International Association of Oil and Gas Producers (IOGP). EPSG code adalah standar internasional yang digunakan dalam berbagai aplikasi geospasial, seperti sistem informasi geografis (GIS), pemetaan, dan navigasi. Berikut adalah penjelasan lebih rinci tentang EPSG code:

1. **Nomor Identifikasi Unik**: Setiap EPSG code adalah nomor identifikasi unik yang merujuk pada sistem referensi koordinat tertentu. Nomor ini sering digunakan dalam format "EPSG:xxxx", di mana "xxxx" adalah nomor identifikasi unik.

2. **Deskripsi Sistem Referensi Koordinat**: Setiap EPSG code memiliki deskripsi yang menjelaskan sistem referensi koordinat yang terkait. Ini termasuk informasi tentang jenis proyeksi peta, datum geodetik, unit pengukuran, dan parameter lain yang diperlukan untuk melakukan transformasi koordinat.

3. **Keuntungan Universal**: EPSG code memiliki keuntungan universal karena mereka diterima dan digunakan di seluruh dunia. Ini membuatnya sangat berguna dalam pertukaran data geografis antar berbagai perangkat lunak dan sistem geospasial.

4. **Mendukung Berbagai Jenis CRS**: EPSG code mencakup berbagai jenis sistem referensi koordinat, termasuk sistem geografis (latitude/longitude), proyeksi peta, sistem projeksi khusus, dan datum geodetik. Dengan demikian, Anda dapat menemukan EPSG code untuk berbagai keperluan.

5. **Integrasi dalam Perangkat Lunak**: Banyak perangkat lunak geospasial dan GIS telah mengintegrasikan daftar EPSG code ke dalam sistem mereka. Ini memungkinkan pengguna untuk dengan mudah memilih sistem referensi koordinat yang sesuai saat membuat atau mengimpor data geografis.

Contoh EPSG code yang umum adalah:

    EPSG:4326 untuk sistem referensi koordinat geografis WGS 84 (latitude/longitude).
    EPSG:3857 untuk sistem proyeksi peta Web Mercator.
    
Dengan menggunakan EPSG code, Anda dapat dengan mudah merujuk dan berkomunikasi tentang sistem referensi koordinat tertentu tanpa kebingungan. Hal ini sangat penting dalam pemetaan, analisis geografis, dan berbagai aplikasi yang melibatkan data geografis.

## Why Converting the CRS ?`

Konversi Sistem Referensi Koordinat (Coordinate Reference System atau CRS) diperlukan dalam berbagai konteks geospasial dan pemetaan karena alasan berikut:

- **Kompabilitas Data**: Data geografis seringkali diterbitkan atau dikumpulkan dalam berbagai sistem referensi koordinat. Untuk menggabungkan, menganalisis, atau membandingkan data dari sumber yang berbeda, Anda perlu mengonversi data ke CRS yang konsisten.

- **Keperluan Proyeksi**: Sistem proyeksi peta digunakan untuk merepresentasikan permukaan lengkung Bumi dalam bidang datar. Proyeksi peta yang berbeda memiliki karakteristik yang berbeda, seperti distorsi dan skala. Konversi CRS diperlukan untuk memproyeksikan data ke sistem proyeksi peta tertentu yang sesuai dengan tujuan analisis atau visualisasi.


- **Analisis Spasial**: Analisis geografis seringkali memerlukan data dalam format yang seragam. Misalnya, ketika Anda melakukan analisis overlay (overlaying) untuk mengidentifikasi interaksi spasial antara berbagai lapisan data, semua data harus berada dalam CRS yang sama.

- **Navigasi dan Navigasi GPS**: Sistem navigasi, seperti GPS (Global Positioning System), menggunakan sistem referensi koordinat tertentu. Untuk menggunaan data GPS dalam aplikasi yang memerlukan sistem referensi koordinat yang berbeda, Anda perlu mengonversinya.

- **Pemetaan Topografis**: Dalam survei topografis dan pemetaan lahan, data harus dipetakan dalam sistem referensi koordinat yang sesuai dengan sistem pemetaan yang digunakan.

- **Standar Industri**: Beberapa industri memiliki standar penggunaan sistem referensi koordinat tertentu. Misalnya, industri perminyakan dan gas sering menggunakan sistem referensi koordinat tertentu dalam eksplorasi dan pemetaan sumber daya.

- **Presentasi Data**: Ketika Anda ingin menghasilkan peta atau visualisasi data geografis, sistem referensi koordinat yang dipilih dapat memengaruhi tampilan akhir dan pemahaman data.

- **Perubahan Zona Waktu**: Konversi CRS juga mungkin diperlukan dalam konteks perubahan zona waktu. Beberapa data geografis menyertakan informasi zona waktu yang berbeda, dan konversi dapat diperlukan untuk konsistensi.

- **Perubahan Datum**: Data yang dikumpulkan dengan datum geodetik tertentu mungkin perlu diubah ke datum yang berbeda untuk akurasi yang lebih baik dalam analisis geodetik dan geospasial.

Konversi CRS adalah langkah penting dalam pemrosesan dan analisis data geografis. Ini memungkinkan data yang berasal dari berbagai sumber dan dalam berbagai format untuk digunakan bersama dan untuk tujuan analisis yang spesifik. Pemahaman tentang konsep ini sangat penting dalam bidang pemetaan, geospasial, dan ilmu geografi.

### How to choose which CRS to use ?

Memilih Coordinate Reference System (CRS) yang tepat adalah langkah penting dalam pemetaan, analisis geografis, dan pemrosesan data geografis. Berikut adalah langkah-langkah untuk memilih CRS yang sesuai:

**1. Tentukan Tujuan Anda**: Pertama, tentukan dengan jelas tujuan Anda. Apakah Anda ingin membuat peta, melakukan analisis geografis, navigasi, pemodelan lingkungan, atau tujuan lainnya? Tujuan Anda akan memengaruhi pilihan CRS.

**2. Pahami Data Anda**: Ketahui sistem referensi koordinat yang digunakan oleh data Anda. Informasi CRS data biasanya tersedia dalam metadatanya. Pastikan data Anda memiliki CRS yang sesuai dengan tujuan Anda.

**3. Ketahui Area Studi Anda:** Pertimbangkan ukuran dan lokasi area studi Anda. Jika Anda bekerja dalam skala lokal, mungkin lebih baik menggunakan sistem referensi koordinat yang cocok dengan area tersebut. Untuk proyek yang melibatkan area besar, Anda mungkin memilih proyeksi peta yang lebih cocok untuk seluruh wilayah.

**4. Perhatikan Distorsi:** Berbagai proyeksi peta dapat menghasilkan distorsi yang berbeda, seperti distorsi sudut, area, atau jarak. Pilih proyeksi yang paling sesuai dengan jenis distorsi yang dapat diterima dalam analisis Anda.

**5. Ketahui Datum:** Pastikan data Anda menggunakan datum yang sesuai dengan lokasi Anda. Misalnya, di Amerika Utara, North American Datum (NAD) biasa digunakan, sementara WGS 84 adalah datum global yang umum.

**6. Perhatikan Standar Industri:** Beberapa industri memiliki standar penggunaan CRS tertentu. Misalnya, industri perminyakan dan gas sering menggunakan proyeksi peta tertentu untuk pemetaan sumber daya.

**7. Gunakan Sumber Referensi:** Gunakan sumber data atau literatur referensi yang bisa membantu Anda memilih CRS yang sesuai. Sumber resmi seperti peta nasional atau lembaga geospasial sering menyediakan informasi tentang CRS yang sesuai.

**8. Konsultasikan dengan Ahli:** Jika Anda tidak yakin, konsultasikan dengan ahli geospasial, pemetaan, atau GIS. Mereka memiliki pengetahuan mendalam tentang CRS dan dapat membantu Anda memilih yang sesuai.

**9. Coba dan Evaluasi:** Jika memungkinkan, coba beberapa CRS yang mungkin cocok untuk Anda, lalu evaluasi hasilnya. Pilih yang memberikan hasil terbaik sesuai dengan tujuan Anda.

**10. Dokumentasikan Pilihan Anda:** Setelah Anda memilih CRS, pastikan untuk mendokumentasikan pilihan Anda. Informasikan kepada orang lain yang mungkin akan bekerja dengan data tersebut tentang CRS yang digunakan.

**11. Perhatikan Format Data:** Periksa juga format data yang Anda gunakan. Beberapa format data (seperti shapefile atau GeoJSON) mungkin sudah mencantumkan informasi tentang CRS, sehingga Anda perlu memahami cara bekerja dengan format tersebut.

Ingatlah bahwa pemilihan CRS yang tepat dapat sangat memengaruhi akurasi analisis dan visualisasi data geografis Anda. Dengan pemahaman yang baik tentang tujuan Anda, data Anda, dan karakteristik area studi, Anda dapat membuat pilihan yang tepat dalam memilih CRS yang sesuai.

**Projecting a Point**

In the previous chapter, we worked with the Eiffel Tower location. Again, we provided you the coordinates in a projected coordinate system, so you could, for example, calculate distances. Let's return to this iconic landmark, and express its location in geographical coordinates: 48°51′29.6″N, 2°17′40.2″E. Or, in decimals: latitude of 48.8584 and longitude of 2.2945.

Shapely geometry objects have no notion of a CRS, and thus cannot be directly converted to another CRS. Therefore, we are going to use the GeoPandas to transform the Eiffel Tower point location to an alternative CRS. We will put the single point in a GeoSeries, use the to_crs() method, and extract the point again.

GeoPandas is already imported.

- Create a shapely point object with the coordinates of the Eiffel Tower and assign it to a variable called eiffel_tower.
- Create a GeoSeries (called s_eiffel_tower) with the Eiffel Tower as the single element and specify the CRS to be EPSG:4326.
- Convert s_eiffel_tower to EPSG:2154, and call the result s_eiffel_tower_projected

```python
# Construct a Point object for the Eiffel Tower
from shapely.geometry import Point
eiffel_tower = Point(2.2945, 48.8584)
```

```python
# Put the point in a GeoSeries with the correct CRS
s_eiffel_tower = geopandas.GeoSeries([eiffel_tower], crs={'init': 'epsg:4326'})
```

```python
# Convert to other CRS
s_eiffel_tower_projected = s_eiffel_tower.to_crs(epsg=2154)
```

```python
# Print the projected point
print(s_eiffel_tower_projected)
```

> That did it! It's a little bit cumbersome to project a single Point like that, but now we are ready to use this location with the other Paris datasets.

## Spatial Operations 

Spatial operations (operasi spasial) merujuk pada serangkaian tindakan dan perhitungan yang dilakukan pada data geografis atau geometri dalam konteks geospasial atau sistem informasi geografis (GIS). Operasi spasial memungkinkan pengguna untuk melakukan analisis, perbandingan, pemodelan, dan visualisasi data geografis. Berikut beberapa contoh operasi spasial yang umum:

**1. Intersection**

Intersect adalah operasi yang menghasilkan geometri poligon baru yang merupakan perpotongan dari dua geometri yang diberikan. Ini sering digunakan untuk menemukan titik perpotongan antara fitur-fitur di dua lapisan data geografis.


**2. Union**

Operasi penggabungan (union) menggabungkan dua atau lebih geometri menjadi satu geometri tunggal. Ini digunakan untuk menggabungkan wilayah yang tumpang tindih atau untuk menghasilkan wilayah gabungan dari beberapa fitur.


**3. Difference**

menghitung perbedaan geometri antara dua atau lebih lapisan data geografis. Operasi ini menghasilkan geometri yang merupakan perbedaan atau perpotongan antara geometri yang diberikan. Misalnya, ketika menjalankan operasi difference pada dua polygon, hasilnya adalah polygon yang mewakili area yang hanya ada di salah satu polygon tetapi tidak ada di yang lain.
 


### Spatial Operation in Geopandas 

```python
africa = geopandas.read_file('dataset/africa.geojson')
africa.head()
```

```python
from shapely.geometry import LineString
box = LineString([(-10, 0), (50, 0)]).buffer(10, cap_style=3)
print(box)
```

# Overlaying Spatial Dataset

Dalam analisis data geospasial di Python, "overlay" dan "intersection" adalah dua operasi yang digunakan untuk memanipulasi dan menganalisis geometri. Berikut adalah perbedaan antara kedua operasi ini:

**Overlay:**

- Operasi overlay adalah operasi yang menggabungkan atau menggabungkan geometri dari dua atau lebih lapisan data geografis. Ini biasanya menghasilkan data gabungan yang mempertahankan fitur-fitur dari lapisan asal.
- Operasi overlay dapat digunakan untuk mengidentifikasi area tumpang tindih antara lapisan, menggabungkan fitur yang tumpang tindih, atau membuat representasi gabungan dari beberapa lapisan data geografis.
- Di Python, perpustakaan geospasial seperti GeoPandas menyediakan fungsi overlay() yang memungkinkan Anda untuk melakukan operasi overlay dengan mudah.

**Intersection:**

- Operasi intersection (perpotongan) adalah operasi yang menghasilkan geometri yang merupakan area yang tumpang tindih antara dua atau lebih geometri. Ini mengidentifikasi area di mana geometri dari lapisan data geografis bersilangan.
- Operasi ini berguna untuk mengidentifikasi perpotongan spasial antara entitas, menghitung luas area tumpang tindih, atau menggabungkan atribut dari entitas yang bersilangan.
- Di Python, perpustakaan geospasial seperti GeoPandas menyediakan fungsi intersection() yang memungkinkan Anda untuk melakukan operasi perpotongan dengan mudah.

**Perbedaan utama antara kedua operasi ini adalah output yang dihasilkan:**

- Operasi overlay menghasilkan data yang mempertahankan fitur-fitur dari lapisan asal dengan modifikasi yang mungkin tergantung pada jenis overlay yang digunakan (misalnya, union atau intersection).
- Operasi intersection menghasilkan geometri yang hanya mencakup area yang bersilangan antara geometri dari lapisan yang diberikan.

Kedua operasi ini berguna dalam analisis data geospasial, dan pemilihan operasi tergantung pada tujuan analisis Anda. Jika Anda ingin menggabungkan geometri atau mengidentifikasi area bersilangan, overlay mungkin lebih sesuai. Jika Anda hanya ingin mengidentifikasi area bersilangan tanpa menggabungkan fitur, maka intersection adalah pilihan yang tepat.
