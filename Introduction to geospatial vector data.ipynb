# Introduction to Geospatial Vector Data

Data Geospasial adalah data yang terkait dengan lokasi geografis dan informasi spasial.  Ini mencakup berbagai informasi yang dihubungkan dengan koordinat geografis, seperti data tentang bumi, permukaan laut, bangunan, jalan, dan wilayah geografis lainnya. Data geospasial sangat penting dalam berbagai aplikasi, termasuk pemetaan, pemantauan lingkungan, manajemen sumber daya alam, navigasi, perencanaan kota, dan banyak lagi 


**Komponen Data Geospasial:**

- Data Geografis: Ini mencakup informasi tentang lokasi geografis seperti koordinat lintang dan bujur (Latitude and Longitude), kontur tanah, sungai, dan lain-lain.
- Data Non-Geografis: Ini adalah data yang terkait dengan entitas geografis, seperti informasi yang ada di dalam database yang mengacu pada lokasi tertentu.

**Jenis Data Geospasial:**

- Vektor: Data vektor menggambarkan objek geografis sebagai kombinasi dari titik, garis, dan poligon. Contoh termasuk peta jalan, batas wilayah administratif, dan lokasi titik seperti bangunan.
- Raster: Data raster digambarkan sebagai grid atau kisi piksel dan digunakan untuk menggambarkan data kontinu seperti citra satelit, data cuaca, dan citra penginderaan jauh.

**Sumber Data Geospasial:**

- Pemetaan Tradisional: Data pemetaan tradisional dikumpulkan melalui survei lapangan dengan menggunakan perangkat seperti GPS (Global Positioning System).
- Citra Satelit: Citra satelit yang diambil dari satelit ruang angkasa adalah sumber data geospasial penting untuk pemetaan dan pemantauan permukaan bumi.
- Penginderaan Jauh: Teknologi seperti lidar, radar, dan fotogrametri digunakan untuk mengumpulkan data geospasial dengan cara yang lebih canggih.

**Manfaat Data Geospasial:**

- Pemetaan dan Navigasi: Data geospasial digunakan dalam sistem GPS dan peta digital untuk navigasi dan mencari lokasi.
- Perencanaan Kota dan Pengembangan Wilayah: Data ini membantu pemerintah dan perusahaan swasta dalam perencanaan perkotaan, pemetaan zona penggunaan lahan, dan infrastruktur kota.
- Pemantauan Lingkungan: Data geospasial membantu dalam pemantauan iklim, perubahan tanah, kualitas air, dan banyak aspek lingkungan lainnya.
- Manajemen Sumber Daya Alam: Data ini digunakan untuk pemantauan dan pengelolaan sumber daya alam seperti hutan, pertanian, dan perikanan.
- Manajemen Bencana: Data geospasial membantu dalam perencanaan bencana, respons darurat, dan pemantauan bencana alam seperti banjir, gempa bumi, dan badai.

**Perangkat Lunak SIG (Sistem Informasi Geografis):**
- Perangkat lunak SIG memungkinkan pengguna untuk mengolah, menganalisis, dan memvisualisasikan data geospasial. Contoh perangkat lunak SIG termasuk ArcGIS, QGIS, dan Google Earth.

Data geospasial memiliki peran penting dalam pemahaman dan pengelolaan dunia di sekitar kita. Dengan teknologi dan alat yang terus berkembang, penggunaan data geospasial akan terus berkembang untuk memecahkan masalah kompleks di berbagai bidang.

3. Polygon, Garis- garis yang saling menutup hingga membentuk sebuah bidang/area yang menghasilkan panjang dan lebar. 

dari 3 jenis diatas, dalam python ditambahkan beberapa geometri baru untuk memudahkan pengolahan data spasial

4. MultiPolygon, kumpulan polygon-polygon

# Introduction to GeoPandas

GeoPandas adalah sebuah pustaka atau perpustakaan (library) open-source di Python yang dirancang untuk mempermudah analisis dan manipulasi data geospasial (data yang terkait dengan lokasi geografis) dengan mudah dan efisien. GeoPandas memadukan beberapa pustaka Python populer seperti Pandas, Matplotlib, Shapely, Fiona, dan lainnya untuk memungkinkan pengguna bekerja dengan data geospasial dengan cara yang mirip dengan cara penggunaan data tabel dalam Pandas. Berikut penjelasan yang lebih komprehensif tentang GeoPandas:

- **Data Geospasial**: GeoPandas memungkinkan Anda untuk bekerja dengan berbagai jenis data geospasial, termasuk data geometri seperti poin (titik), garis (line), dan poligon (polygon), serta data geografis dalam format berkas seperti Shapefiles, GeoJSON, dan lainnya.

- **Penggabungan Data Tabular**: Anda dapat menggabungkan data geospasial dengan data tabular biasa (seperti CSV atau Excel) untuk melakukan analisis lebih mendalam. Ini memungkinkan Anda menghubungkan data atribut (misalnya, nama, populasi, dll.) dengan data geometri (bentuk, lokasi) dalam satu struktur data.

- **Operasi Geometri**: GeoPandas menyediakan banyak operasi geometri yang dapat digunakan untuk memanipulasi data geospasial. Anda dapat melakukan operasi seperti union, intersection, buffer, dissolve, dan lainnya untuk menganalisis dan memanipulasi data geometri.

- **Visualisasi**: Anda dapat dengan mudah membuat visualisasi dari data geospasial menggunakan Matplotlib atau fungsi bawaan GeoPandas untuk membuat peta, plot data geospasial, dan menambahkan informasi lain seperti legenda dan judul.

- **Analisis Geospasial**: Anda dapat melakukan berbagai jenis analisis geospasial, termasuk analisis jarak, analisis klastering spasial, analisis overlay (seperti intersect dan union), dan banyak lagi. Hal ini berguna untuk pemodelan data geografis dan ilmu geografi.

- **Interoperabilitas**: GeoPandas mendukung banyak format berkas geospasial dan dapat digunakan bersamaan dengan pustaka geospasial lainnya seperti GDAL, PROJ, dan PostGIS.

- **Open Source dan Komunitas Aktif**: GeoPandas adalah proyek open-source, yang berarti Anda dapat menggunakannya secara gratis dan berkontribusi dalam pengembangannya. Terdapat komunitas yang aktif yang dapat membantu dalam pertanyaan dan solusi masalah.

GeoPandas sangat berguna untuk berbagai aplikasi yang melibatkan data geografis, seperti pemetaan, analisis lingkungan, perencanaan tata guna lahan, analisis transportasi, ilmu bumi, dan banyak aplikasi lainnya. Dengan kesederhanaan dan kemampuannya yang kuat dalam mengelola data geospasial, GeoPandas telah menjadi alat yang populer dalam analisis geospasial dengan Python.

**The Paris restaurants as a GeoDataFrame**

In the first coding exercise of this chapter, we imported the locations of the restaurants in Paris from a csv file. To enable the geospatial functionality of GeoPandas, we want to convert the pandas DataFrame to a GeoDataFrame. This can be done with the GeoDataFrame() constructor and the geopandas.points_from_xy() function, and is done for you.

Now we have a GeoDataFrame, all spatial functionality becomes available, such as plotting the geometries. In this exercise we will make the same figure as in the first exercise with the restaurants dataset, but now using the GeoDataFrame's plot() method.

Pandas has been imported as pd, GeoPandas as geopandas and matplotlib's pyplot functionality as plt

- Inspect the first rows of the restaurants GeoDataFrame.
- Plot it with the plot() method of the GeoDataFrame. The return value is a matplotlib axes object: call this ax.
- Again, make sure to set the marker size equal to 1.
- Add a basemap layer using contextily.

```python
# Read the restaurants csv file into a DataFrame
df = pd.read_csv("dataset/paris_restaurants.csv")

# Convert it to a GeoDataFrame
restaurants = geopandas.GeoDataFrame(df, geometry=geopandas.points_from_xy(df.x, df.y))

# Inspect the first rows of the restaurants GeoDataFrame
print(restaurants.head())

# Make a plot of the restaurants
ax = restaurants.plot(markersize=1)
import contextily
contextily.add_basemap(ax)
plt.show()
```

> We now have the same plot as we constructed before with the normal pandas DataFrame! Do you notice the difference in how to make it? This time, we used the specialized plot() method of the GeoDataFrame, and we didn't need to specify the x and y coordinates explicitly.

**Visualizing the population density**

Let's get back to the districts dataset. In a previous exercise we visualized the districts with a uniform column. But often we want to show the spatial variation of a variable, and color the polygons accordingly.

In this exercise we will visualize the spatial variation of the population density within the center of Paris. For this, we will first calculate the population density by dividing the population number with the area, and add it as a new column to the dataframe.

The districts dataset is already loaded as districts, GeoPandas has been imported as geopandas and matplotlib.pyplot as plt.

- Print the first rows of the districts dataset. Do you see the 'population' column?
- Inspect the area of the districts's geometries.
- Add a column 'population_density' representing the number of inhabitants per squared kilometer (Note: The area is given in squared meter, so you will need to multiply the result with 10**6).
- Plot the districts using the 'population_density' to color the polygons.

```python
# Inspect the first rows of the districts dataset
print(districts.head())

# Inspect the area of the districts
print(districts.geometry.area)

# Add a population density column
districts['population_density'] = districts['population'] / districts.geometry.area * 10**6

# Make a plot of the districts colored by the population density
districts.plot(column='population_density', legend=True)
plt.show()
```

> Nice work! The column keyword provides an easy way to make such plots. In a later chapter, we will also see how to customize the color scale.

**Using pandas functionality: groupby**
    
This course will focus on the spatial functionality of GeoPandas, but don't forget that we still have a dataframe, and all functionality you know from Pandas is still applicable.

In this exercise, we will recap a common functionality: the groupby operation. You may want to use this operation when you have a column containing groups, and you want to calculate a statistic for each group. In the groupby() method, you pass the column that contains the groups. On the resulting object, you can then call the method you want to calculate for each group. In this exercise, we want to know the size of each group of type of restaurants.

We refer to the course on Manipulating DataFrames with pandas for more information and exercises on this groupby operation.

- Using groupby(), group the restaurants by the type of restaurants, and calculate the size of each group. Call the result type_counts.
- Print the resulting Series.

```python
# Load the restaurants data
restaurants = geopandas.read_file("dataset/paris_restaurants.csv")

# Calculate the number of restaurants of each type
type_counts = restaurants.groupby('type').size()

# Print the result
print(type_counts)
```

> So the traditional French restaurants are the most common ones. We wanted to show you the groupby functionality here, but note that for counting values in a column, there is also the value_counts() method that does the same.
